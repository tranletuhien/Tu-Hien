```python
import pandas as pd
url='passenger1.csv'
data = pd.read_csv(url)
data.head()
```

```python

import seaborn as sns
import matplotlib.pyplot as plt
data.hist()
plt.show()
```

```python
correlation=data.corr()
print(correlation)
```

```python
sns.pairplot(data)
```

```python
sns.heatmap(correlation,annot=True)
```

```python
data.plot(kind='scatter',x='Passenger',y='Dollars (Y)',figsize=(9,9),color='black')
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['Passenger']),y=data['Dollars (Y)'])
print('Giá trị Intercept là:',model.intercept_)
print('Giá trị Coeffient là:',model.coef_)
```

```python
model.score(X=pd.DataFrame(data['Passenger']),y=data['Dollars (Y)'])
```

```python
data.plot(kind='scatter',x='Passenger',y='Dollars (Y)',figsize=(9,9),color='black')
y_pred=model.predict(X=pd.DataFrame(data['Passenger']))
plt.plot(data['Passenger'],y_pred,color='blue')
```

```python
model=linear_model.LinearRegression()
X=pd.DataFrame([data['Passenger'],data['Passenger']**2]).T
model.fit(X,y=data['Dollars (Y)'])
print('Hệ số Intercept là:',model.intercept_)
print('Hệ số Coeffient là:',model.coef_)
print('Độ chính xác của mô hình:',model.score(X,y=data['Dollars (Y)']))
```
